```python
import matplotlib.pyplot as plt

time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]

# 1
low_study = []      # less than 3 hours
moderate_study = [] # between 3 and 5 hours.
high_study = []     # more than 5 hours

# 2.
for day_data in time_data:
    study_hours = day_data[0]

    if study_hours < 3:
        low_study.append(study_hours)
    elif 3 <= study_hours <= 5:
        moderate_study.append(study_hours)
    else:  # study_hours > 5
        high_study.append(study_hours)

# 3.
print("Low Study Time (less than 3 hours):", low_study)
print("Moderate Study Time (3to5 hours):", moderate_study)
print("High Study Time (more than 5 hours):", high_study)

```

```python
# recursive
# Task1
def sum_nested_list(nested_list):

    total = 0

    for element in nested_list:
        if isinstance(element, list):

            total += sum_nested_list(element)
        else:

            total += element

    return total

# Test the function with the sample nested list
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)

print(f"Sample nested list: {nested_list}")
print(f"Total sum: {result}")
```

```python
# Task2

def generate_permutations(s):
    if len(s) <= 1:
        return [s]
    permutations = []
    for i, char in enumerate(s):
        remaining_chars = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining_chars):
            permutations.append(char + perm)

    return permutations

def generate_permutations_unique(s):
    if len(s) <= 1:
        return [s]

    permutations = set()

    for i, char in enumerate(s):
        remaining_chars = s[:i] + s[i+1:]
        for perm in generate_permutations_unique(remaining_chars):
            permutations.add(char + perm)

    return list(permutations)

# Test the functions
print("Testing with 'abc':")
result_abc = generate_permutations("abc")
print(result_abc)
print(f"Number of permutations: {len(result_abc)}")

print("\nTesting with 'aab' (using basic function - may have duplicates):")
result_aab_basic = generate_permutations("aab")
print(result_aab_basic)
print(f"Number of permutations: {len(result_aab_basic)}")

print("\nTesting with 'aab' (using unique function - no duplicates):")
result_aab_unique = generate_permutations_unique("aab")
print(result_aab_unique)
print(f"Number of unique permutations: {len(result_aab_unique)}")
```

```python
# Task3

def calculate_directory_size(directory):
    total_size = 0

    for key, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            total_size += value

    return total_size

# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

# Test the function
total_size = calculate_directory_size(directory_structure)
print(f"Total directory size: {total_size} KB")
```

```python
# Dynamic programming
# Task1

def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Test with coins = [1, 2, 5] and amount = 11
coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
print(f"Minimum coins needed: {result}")
```

```python
# Task2
def longest_common_subsequence(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]

# Test with strings "abcde" and "ace"
s1 = "abcde"
s2 = "ace"
result = longest_common_subsequence(s1, s2)
print(f"String 1: '{s1}'")
print(f"String 2: '{s2}'")
print(f"Length of Longest Common Subsequence: {result}")
```

```python
# Task3

def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i-1] <= w:
                dp[i][w] = max(dp[i-1][w], values[i-1] + dp[i-1][w - weights[i-1]])
            else:

                dp[i][w] = dp[i-1][w]

    return dp[n][capacity]

# Test with given values
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
result = knapsack(weights, values, capacity)
print(f"Weights: {weights}")
print(f"Values: {values}")
print(f"Capacity: {capacity}")
print(f"Maximum value: {result}")
```
