3.2 Exploratory Data Analysis – Exercise 1
Fill Missing Age Values (required step)

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

Question 1 – Highest fare relative to age
1. Add fare_per_year

```python
df["fare_per_year"] = df["Fare"] / df["Age"]
```

2. Subset fare_per_year > 5

```python
high_fare_age = df[df["fare_per_year"] > 5]
```

3. Sort descending

```python
high_fare_age_srt = high_fare_age.sort_values(by="fare_per_year", ascending=False)
```

4. Select Name and fare_per_year

```python
result = high_fare_age_srt[["Name", "fare_per_year"]]
result
```

Question 2 – Adult male, highest fare relative to class
1. Add fare_per_class

```python
df["fare_per_class"] = df["Fare"] / df["Pclass"]
```

2. Subset adult males

```python
adult_males = df[(df["Sex"] == "male") & (df["Age"] >= 18)]
```

3. Sort descending

```python
adult_males_srt = adult_males.sort_values(by="fare_per_class", ascending=False)
```

4. Select Name, Age, fare_per_class

```python
result = adult_males_srt[["Name", "Age", "fare_per_class"]]
result
```

3.3 Group-By / Percentage Questions
Question 1 – Percent of total fare revenue by class
1. Total fare revenue

```python
total_fare = df["Fare"].sum()
```

2. Fare by class

```python
fare_class_1 = df[df["Pclass"] == 1]["Fare"].sum()
fare_class_2 = df[df["Pclass"] == 2]["Fare"].sum()
fare_class_3 = df[df["Pclass"] == 3]["Fare"].sum()
```

3. Convert to percentage

```python
fare_percent_1 = (fare_class_1 / total_fare) * 100
fare_percent_2 = (fare_class_2 / total_fare) * 100
fare_percent_3 = (fare_class_3 / total_fare) * 100

fare_percent_1, fare_percent_2, fare_percent_3
```

Question 2 – Percent of passengers by age group
1. Create age_group column

```python
df["age_group"] = "adult"
df.loc[df["Age"] < 18, "age_group"] = "child"
df.loc[df["Age"] >= 65, "age_group"] = "senior"
```

2. Total passengers

```python
total_passengers = len(df)
```

3. Count each age group

```python
age_group_counts = df["age_group"].value_counts()
```

4. Convert to percentage

```python
age_group_percent = (age_group_counts / total_passengers) * 100
age_group_percent
```
